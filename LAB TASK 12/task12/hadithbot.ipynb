```python
import glob
path = '/content/LK-Hadith-Corpus'
files = sorted(glob.glob(path + '//**//*.csv',recursive = True))
print(files)
```

```python
import re

def clean_text(text):
  text = text.lower()
  text = re.sub(r'[^a-zA-Z0-9\s]','',text)
  text = re.sub(r'[\s+]','',text)
  return text
```

```python
columns = [
    'Chapter_Number' , 'Chapter_English' ,'Chapter_Arabic', 'Section_Number',
    'Section_English', 'Section_Arabic', 'Hadith_Number',
    'English_Hdith','English_Hadith','English_Isnad','English_Matn','Arabic_Hadith'
    'Arabic_Grade'
]
columns.append('Cleaned_Hadith')
```

```python
import pandas as pd
all_hadith = []
for file in files:
  df = pd.read_csv(file,names = columns,skiprows=1)
  df['Cleaned_Hadith'] = df['English_Hadith'].astype(str).apply(clean_text)
  all_hadith.extend(df[columns].values.tolist())
```

```python
hadith_df = pd.DataFrame(all_hadith, columns = columns)
hadith_df.head()
```

```python
hadith_df.to_csv('cleaned_hadith.csv', index = False)
```

```python
import faiss
dimensions = embeddings.shape[1]
faiss_index = faiss.IndexFlatL2(dimensions)
faiss_index.add(embeddings)
faiss.write_index(faiss_index,'fiasss_index.index')
```

```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')

def get_similiar_hadith(query,count = 5,model = model, fiass_index=faiss_index):
  query_embeddings = model.encode([query])
  distances, indices = fiass_index.search(query_embeddings,count)
  for i in range(count):
    print(hadith_df.iloc[indices[0][i]]['English_Hadith'])
```
