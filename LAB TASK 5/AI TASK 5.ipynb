### __IMPORT LIBARARIES__

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

### __READ & DISPLAY__

```python

img = cv2.imread('hw.png')

if img is None:
    print("Error: Image not found!")
else:
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    plt.imshow(img_rgb)
    plt.title("Watch logo")
    plt.axis('off')
```

### __RESIZED IMAGE__

```python
resized = cv2.resize(img, (200, 300))

plt.imshow(cv2.cvtColor(resized, cv2.COLOR_BGR2RGB))
plt.title("Resized Image")
plt.axis('off')
```

### __CROPPED IMAGE__

```python
crop = img[100:150, 200:300]

plt.imshow(cv2.cvtColor(crop, cv2.COLOR_BGR2RGB))
plt.title("Cropped Image")
plt.axis('off')
```

### __DRAW SHAPE__

```python
rect = img.copy()
cv2.rectangle(rect, (100, 50), (150, 200), (255, 0, 0), 2)

plt.imshow(cv2.cvtColor(rect, cv2.COLOR_BGR2RGB))
plt.title("rectangle")
plt.axis('off')
```

### __ADD TEXT__

```python
text_img = img.copy()
cv2.putText(text_img, 'This is the logo of watches store', (50, 50),
            cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)

plt.imshow(cv2.cvtColor(text_img, cv2.COLOR_BGR2RGB))
plt.title("Text added")
plt.axis('off')
```

### __BLUR IMAGE__

```python
blur = cv2.GaussianBlur(img, (9, 9), 0)

plt.imshow(cv2.cvtColor(blur, cv2.COLOR_BGR2RGB))
plt.title("Blur Image")
plt.axis('off')
```
