## __WATER JUG PROBLEM__

```python

def water_jug_dfs(cap_a, cap_b, target):
    visited= set()  
    stack = [(0, 0, [])]  

    while stack:
        
        a, b, moves = stack.pop()
        
        if a == target or b == target:
            print("solution possible")
            for step in moves:
                print(step)
            print(f"Final State of jug: ({a}, {b})")
            return

        if (a, b) in visited:
            continue
        visited.add((a, b))

    
        pos_moves = []

        # 1. Fill Jug A
        pos_moves.append((cap_a, b, moves + ["fill jug a"]))
        # 2. Fill Jug B
        pos_moves.append((a, cap_b, moves + ["fill jug b"]))
        # 3. Empty Jug A
        pos_moves.append((0, b, moves + ["empty jug a"]))
        # 4. Empty Jug B
        pos_moves.append((a, 0, moves + ["empty jug b"]))
        # 5. Pour A -> B
        pour = min(a, cap_b - b)
        pos_moves.append((a - pour, b + pour, moves + [f"pour {pour} from a to b"]))
        # 6. Pour B -> A
        pour = min(b, cap_a - a)
        pos_moves.append((a + pour, b - pour, moves + [f"pour {pour} from b to a"]))

        stack.extend(pos_moves)

    print("No solution found.")

# Jug A = 4, Jug B = 3, Target = 2
water_jug_dfs(4, 3, 2)
```
